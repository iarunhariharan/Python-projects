```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data

df= pd.read_excel("G:/My Drive/DSML raw files/python/modend.xlsx")

# Display data
print(df)
```

```python
# Replacing values in height with random numbers between 150 and 180

df['Height'] = np.random.randint(150, 181, size=len(df))

# Display data after height modification
print(df)
```

```python
# Count the number of employees in each team
team_distribution = df["Team"].value_counts()

# Calculate the percentage split
team_percentage = (team_distribution / len(df)) * 100

# Distribution
plt.figure(figsize=(12, 6))
sns.barplot(x=team_distribution.index, y=team_distribution.values, palette="rocket")
plt.xticks(rotation=90)
plt.xlabel("Team")
plt.ylabel("Number of Employees")
plt.title("Distribution of Employees Across Teams")
plt.show()

# Display percentage split
team_percentage
```

```python
# Count the number of employees in each position
position_distribution = df["Position"].value_counts()

# Distribution
plt.figure(figsize=(10, 5))
sns.barplot(x=position_distribution.index, y=position_distribution.values, palette="deep")
plt.xlabel("Position")
plt.ylabel("Number of Employees")
plt.title("Distribution of Employees by Position")
plt.show()

# Display position distribution
position_distribution
```

```python
# Create age bins
age_bins = [18, 24, 30, 36, 42, 50]
age_labels = ["18-24", "25-30", "31-36", "37-42", "43-50"]
df["Age Group"] = pd.cut(df["Age"], bins=age_bins, labels=age_labels, right=False)

# Counting the number of employees in each age group
age_distribution = df["Age Group"].value_counts().sort_index()

# Age group distribution
plt.figure(figsize=(8, 5))
sns.barplot(x=age_distribution.index, y=age_distribution.values, palette="cubehelix")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.title("Predominant Age Group Among Employees")
plt.show()

# Display age group distribution
age_distribution
```

```python
# Group by Team and sum the Salary column
team_salary_expenditure = df.groupby("Team")["Salary"].sum().sort_values(ascending=False)

# Group by Position and sum the Salary column
position_salary_expenditure = df.groupby("Position")["Salary"].sum().sort_values(ascending=False)

# Salary expenditure byTeam 
plt.figure(figsize=(12, 6))
sns.barplot(x=team_salary_expenditure.index, y=team_salary_expenditure.values, palette="flare")
plt.xticks(rotation=90)
plt.xlabel("Team")
plt.ylabel("Total Salary Expenditure")
plt.title("Total Salary Expenditure by Team")
plt.show()

# salary expenditure by Position
plt.figure(figsize=(8, 5))
sns.barplot(x=position_salary_expenditure.index, y=position_salary_expenditure.values, palette="Spectral")
plt.xlabel("Position")
plt.ylabel("Total Salary Expenditure")
plt.title("Total Salary Expenditure by Position")
plt.show()

# Display top results
team_salary_expenditure.head(), position_salary_expenditure.head()
```

```python
# Visualizing correlation between Age and Salary
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Age"], y=df["Salary"], alpha=0.6, color="cyan")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Correlation Between Age and Salary")
plt.show()

# Calculate correlation coefficient
age_salary_correlation = df["Age"].corr(df["Salary"])
age_salary_correlation
```
